```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Product_ID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Product_Name": ["PainRelief", "ColdCure", "HeadacheFix", "PainRelief", "ColdCure", "HeadacheFix", 
                     "PainRelief", "ColdCure", "HeadacheFix", "PainRelief"],
    "Region": ["North America", "Europe", "Asia", "North America", "Europe", "Asia", 
               "North America", "Europe", "Asia", "North America"],
    "Marketing_Spend": [50000, 30000, 45000, 52000, 28000, 48000, 51000, 31000, 47000, 53000],
    "Sales": [100000, 60000, 80000, 110000, 55000, 90000, 105000, 65000, 85000, 115000],
    "Effectiveness": [85, 75, 80, 88, 70, 82, 86, 78, 80, 90],
    "Side_Effects": [5, 2, 3, 4, 1, 3, 5, 2, 4, 6],
    "Age_Group": ["20-30", "30-40", "40-50", "50-60", "30-40", "40-50", "20-30", "30-40", "50-60", "20-30"],
    "Trial_Period": ["6 months", "3 months", "6 months", "12 months", "3 months", "6 months", 
                     "6 months", "3 months", "12 months", "6 months"]
}
df = pd.DataFrame(data)

print("Missing values:\n", df.isnull().sum())

df.drop_duplicates(inplace=True)

avg_effectiveness = df.groupby(["Product_Name", "Region"])["Effectiveness"].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_effectiveness, x="Product_Name", y="Effectiveness", hue="Region", palette="Set2")
plt.title("Average Drug Effectiveness by Region")
plt.xlabel("Product Name")
plt.ylabel("Average Effectiveness")
plt.legend(title="Region")
plt.show()

plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x="Product_Name", y="Effectiveness", palette="coolwarm", inner="point")
plt.title("Distribution of Effectiveness by Product")
plt.xlabel("Product Name")
plt.ylabel("Effectiveness")
plt.show()

plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x="Product_Name", y="Side_Effects", palette="coolwarm", inner="point")
plt.title("Distribution of Side Effects by Product")
plt.xlabel("Product Name")
plt.ylabel("Side Effects")
plt.show()

sns.pairplot(df[["Effectiveness", "Side_Effects", "Marketing_Spend"]], kind="scatter", diag_kind="kde", palette="husl")
plt.suptitle("Relationships Between Effectiveness, Side Effects, and Marketing Spend", y=1.02)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Trial_Period", y="Effectiveness", palette="Set3")
plt.title("Effectiveness Across Trial Periods")
plt.xlabel("Trial Period")
plt.ylabel("Effectiveness")
plt.show()

plt.figure(figsize=(8, 5))
sns.regplot(data=df, x="Marketing_Spend", y="Effectiveness", scatter_kws={'alpha':0.7}, line_kws={"color":"red"})
plt.title("Relationship Between Marketing Spend and Drug Effectiveness")
plt.xlabel("Marketing Spend")
plt.ylabel("Effectiveness")
plt.show()
```
